```python
import joblib
import os

# Paths relative to your notebook location
MODEL_DIR = "../model"

MODEL_FILE = os.path.join(MODEL_DIR, "final_xgb_model.pkl")
IMPUTER_FILE = os.path.join(MODEL_DIR, "final_imputer.pkl")
FEATURES_FILE = os.path.join(MODEL_DIR, "final_feature_cols.pkl")

print("Model file exists:", os.path.exists(MODEL_FILE))
print("Imputer file exists:", os.path.exists(IMPUTER_FILE))
print("Features file exists:", os.path.exists(FEATURES_FILE))
print("----")

# Load files
model = joblib.load(MODEL_FILE)
imputer = joblib.load(IMPUTER_FILE)
features = joblib.load(FEATURES_FILE)

print("Length of feature_list:", len(features))
print("Imputer expects n_features:", imputer.n_features_in_)
print("----")

print("Feature List:")
for f in features:
    print(f)
```
